```python
import pandas as pd
import numpy as np
from scipy.stats import zscore

def generate_lag_features(df: pd.DataFrame, n: int) -> pd.DataFrame:
    """
    为每个日期生成前 n 日的特征，包括：
    1. 激活设备数的 z-score 序列
    2. 平均 day1 LTV
    3. 累计贡献的 LTV（考虑回流天数）
    
    参数：
    df: 原始 DataFrame，必须包含 ['dt', 'install_deviceid_cnt', 'day1_ltv_sum', ..., 'day7_ltv_sum']
    n: 往前取几天

    返回：
    增加特征后的 DataFrame
    """
    df = df.sort_values('dt').copy()
    df['dt'] = pd.to_datetime(df['dt'])

    # 1. 激活设备数 lag 特征
    for i in range(1, n + 1):
        df[f'lag_{i}_install_cnt'] = df['install_deviceid_cnt'].shift(i)
    
    df['avg_day1_ltv'] = df['day1_ltv_sum'] / df['install_deviceid_cnt'].replace(0, np.nan)

    for i in range(1, n + 1):
        df[f'lag_{i}_avg_day1_ltv'] = df['avg_day1_ltv'].shift(i)

    # 3. 回流贡献 LTV（不同天数映射不同的 dayX_ltv_sum）
    for i in range(1, n + 1):
        ltv_day = min(i + 1, 7)  # 第i天前的窗口，对应 day{i+1}_ltv_sum
        ltv_col = f'day{ltv_day}_ltv_sum'
        df[f'lag_{i}_contrib_ltv'] = df[ltv_col].shift(i) / df['install_deviceid_cnt'].shift(i)

    return df
```

```python
import pandas as pd
import numpy as np

def row_wise_zscore(df: pd.DataFrame, cols) -> pd.DataFrame:
    """
    对指定列按行做 z-score 标准化。

    参数：
    - df: 原始 DataFrame
    - cols: 需要按行标准化的列名列表

    返回：
    - 一个新的 DataFrame，每列为 'zscore_<原列名>'
    """
    zscore_cols = [f'zscore_{col}' for col in cols]

    def zscore_row(row):
        values = row.values.astype(float)
        mask = ~np.isnan(values)
        valid = values[mask]
        if len(valid) <= 1 or np.std(valid) == 0:
            return pd.Series([np.nan] * len(values), index=zscore_cols)
        else:
            normed = (valid - np.mean(valid)) / np.std(valid)
            result = np.full(len(values), np.nan)
            result[mask] = normed
            return pd.Series(result, index=zscore_cols)

    return df[cols].apply(zscore_row, axis=1)
```

```python
import pandas as pd
import numpy as np

def time_based_zscore(df: pd.DataFrame, date_col: str, cols: list, window_days: int = 180) -> pd.DataFrame:
    """
    对每一列按时间窗口滑动进行 z-score 标准化。

    参数：
    - df: 原始 DataFrame，包含日期列
    - date_col: 日期列名，例如 'dt'
    - cols: 需要标准化的列名
    - window_days: 向前回溯的时间窗口（单位：天）

    返回：
    - 原 df 加上每个列对应的 z-score 列（zscore_<列名>）
    """
    df = df.copy()
    df[date_col] = pd.to_datetime(df[date_col])
    df = df.sort_values(date_col)

    result = df.copy()

    for col in cols:
        zscore_list = []
        for i in range(len(df)):
            current_dt = df.iloc[i][date_col]
            start_dt = current_dt - pd.Timedelta(days=window_days)
            window_df = df[(df[date_col] >= start_dt) & (df[date_col] <= current_dt)]

            values = window_df[col].dropna()
            if len(values) < 2 or np.std(values) == 0:
                zscore_list.append(np.nan)
            else:
                mean = values.mean()
                std = values.std()
                z = (df.iloc[i][col] - mean) / std
                zscore_list.append(z)

        result[f'zscore_{col}'] = zscore_list

    return result
```

```python
import matplotlib.pyplot as plt

def plot_line_chart(df, x_col, y_col, title=None, figsize=(12, 6)):
    """
    画出 df 中指定 x_col 和 y_col 的折线图

    参数:
    - df: 输入的 DataFrame
    - x_col: x轴列名（例如日期）
    - y_col: y轴列名（例如某个比值）
    - title: 图表标题（可选）
    - figsize: 图像大小（默认 12x6）

    返回:
    - None，直接显示图像
    """
    plt.figure(figsize=figsize)
    plt.plot(df[x_col], df[y_col], marker='o', linestyle='-')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    if title:
        plt.title(title)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

def plot_multiple_periods_line_chart(df, periods, x_col, y_col, title=None, figsize=(14, 6)):
    """
    画出多个时间段的 y_col 曲线在一张图中

    参数：
    - df: 输入 DataFrame，包含日期和指标列
    - periods: list of (start_date, end_date) 元组
    - x_col: x轴列名（通常是日期 dt）
    - y_col: y轴列名（指标值）
    - title: 图标题
    - figsize: 图大小
    """
    plt.figure(figsize=figsize)
    for start, end in periods:
        sub_df = df[df[x_col].between(start, end)].copy()
        if pd.api.types.is_datetime64_any_dtype(sub_df[x_col]):
            x_vals = sub_df[x_col].dt.strftime('%m-%d')  # 仅画月份日
        else:
            x_vals = sub_df[x_col]
        plt.plot(x_vals, sub_df[y_col], label=f'{start[:4]}年', marker='o')

    plt.xlabel('日期')
    plt.ylabel(y_col)
    if title:
        plt.title(title)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import euclidean_distances

def find_knn_similar_days(df, target_dt, feature_cols, k=5):
    """
    在 df 中找出与 target_dt 的 feature_cols 向量最相似的 k 个日期（排除自身）
    
    参数：
    - df: 包含日期和特征列的 DataFrame
    - target_dt: 目标日期（字符串或 datetime）
    - feature_cols: 所有特征列名（list）
    - k: 返回的近邻个数

    返回：
    - 相似日期的 DataFrame（含 dt 和距离）
    """

    # 确保 target_dt 是 datetime 类型
    df = df.copy()
    df['dt'] = pd.to_datetime(df['dt'])
    target_dt = pd.to_datetime(target_dt)

    # 拿出目标行和特征向量
    target_row = df[df['dt'] == target_dt]
    if target_row.empty:
        raise ValueError(f"日期 {target_dt} 不在 df 中")

    target_vector = target_row[feature_cols].values  # shape: (1, n_features)

    # 剔除目标日期自身
    df_other = df[df['dt'] != target_dt].copy()
    feature_matrix = df_other[feature_cols].values  # shape: (n_samples, n_features)

    # 计算欧氏距离
    distances = euclidean_distances(feature_matrix, target_vector).flatten()

    # 添加距离列
    df_other['distance'] = distances

    # 返回距离最近的 k 个
    nearest_df = df_other.sort_values(by='distance').head(k)
    return nearest_df
```
